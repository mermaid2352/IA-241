# Import COVID-19 Data

```python
%matplotlib inline
```

```python
import pandas
```

```python
df = pandas.read_excel('s3://deckoia241/COVID-19-geographic-disbtribution-worldwide_April_10_2020.xlsx')

df
```

# Which countries have the highest death toll rates?

```python
sum_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']

sum_death_by_country.nlargest(10).plot.bar()
```

# What is the relationship between the number of cases and the number of deaths in Italy?

```python
italy_data = df.loc[df['countriesAndTerritories'] == 'Italy']
italy_data[:10]
```

```python
italy_data.plot.scatter(x="cases", y="deaths")
```

According to the data above, Italy has the highest amount of cases.

# What is the relationship between the number of cases and the number of deaths in the Netherlands?

```python
netherlands_data = df.loc[df['countriesAndTerritories'] == 'Netherlands']

netherlands_data[:10]
```

```python
netherlands_data.plot.scatter(x="cases", y="deaths")
```

To compare, the Netherlands has the ninth highest amount of cases.
